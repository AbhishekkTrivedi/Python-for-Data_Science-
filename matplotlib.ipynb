```
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

Creating Array 'x' for X-axis 

```
x=np.arange(1,11)
x
```

Creating Array 'y' for Y-axis

```
y = 2*x
y
```

To plot our axis 'X' &'Y'

```
plt.plot(x,y)
plt.show()  
```

Adding Title of Graph & X-axis, Y-axis Labels

```
plt.plot(x,y)
plt.title("Sher ka Graph")
plt.xlabel("CARS")
plt.ylabel("Avg.Speed")
plt.show()
```

Changing Line looks

```
plt.plot(x,y,color='g',linestyle=':',linewidth=6)
plt.show()
```

Adding two Lines in same plot

```
x1=np.arange(1,11)
y1=2*x
y2=8+x
```

```
print(x1)
print(y1)
print(y2)
```

Two lines in a single plot with grid


```
plt.plot(x1,y1,color='b',linestyle='-.',linewidth=3)
plt.plot(x1,y2,color='r',linestyle=':',linewidth=4)
plt.title ('sher ka dusra graph')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.grid (True)
plt.show()

```

Adding Sub plots (i.e. making two plots in same screen space)

```
x2 = np.arange(1,11)
y3=4*x2
y4=2*x2

plt.subplot(1,2,1)
plt.plot(x2,y3,color='g',linestyle="-.",linewidth= 3)

plt.subplot(3,2,2)
plt.plot(x2,y3,color='r',linestyle="-.",linewidth= 3)


```

```
plt.subplot(2,2,3)
plt.plot(x2,y4,color='g',linestyle="-.",linewidth= 3)

plt.subplot(3,4,4)
plt.plot(x2,y4,color='g',linestyle="-.",linewidth= 3)
```

Bar Plot

```
Tecx = {"Abhishek":95,"abhi":58,'sher':74,"aman":85}
```

```
names=list(Tecx.keys())
```

```
names
```

```
marks = list(Tecx.values())
```

```
marks
```

```
plt.bar(names,marks)
plt.show()
```

```
plt.bar(names,marks)
plt.title("Result")
plt.xlabel('names of student')
plt.ylabel('marks of student')
plt.grid(True)
plt.show()
```

Horizontal Bar_Plot (use "plt.barh" instead of "plt.bar" )

```
plt.barh(names,marks,color='r')
plt.title("Result")
plt.xlabel('names of student')
plt.ylabel('marks of student')
plt.grid(True)
plt.show()
```

Scatter Plot

```
p = [1,2,3,4,5,6,7,8,9]
q= [56,76,44,67,35,99,85,24,1]
```

```
plt.scatter(p,q)
plt.show()
```

Changing design & size of dots

```
plt.scatter(p,q,marker='*',c='y',s=300)
plt.show()
```

Two Scatter plot in One_Graph

```
p = [1,2,3,4,5,6,7,8,9]
q= [56,76,44,67,35,99,85,24,1]
r= [35,66,43,99,25,49,61,77,84]
```

```
plt.scatter(p,q,marker='*',c='y',s=300)
plt.scatter(p,r,c='r',s=400)
plt.show()
```

Adding Sub_plots for Scatter Graph


```
p = [1,2,3,4,5,6,7,8,9]
q= [56,76,44,67,35,99,85,24,1]
r= [35,66,43,99,25,49,61,77,84]
```

```
plt.subplot(2,2,1)
plt.scatter(p,q,marker='*',c='g',s=200)

plt.subplot(1,2,2)
plt.scatter(p,r,c='r',s=100)

plt.show()
```

Pie-Chart

```
Subject= ['English','Maths','Hindi','Physics','Chemistry']
Result = [82,95,82,79,95]
```

```
plt.pie(Result,labels=Subject)
plt.show()
```

To Add percentage use 'autopct'

```
plt.pie(Result,labels=Subject,autopct='%0.1f%%',
        colors=['grey','blue','black','pink','green'])
plt.show()
```

```
plt.pie(Result,labels=Subject,autopct='%0.6f%%',
        colors=['grey','blue','black','pink','green'])
plt.show()
```

DoughNut-Chart

```
plt.pie(Result,labels=Subject,radius=2)
plt.pie([1],colors=['w'],radius=1)
plt.show()
```
